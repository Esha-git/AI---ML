```python
# Simple arithmetic
print("2 + 3 =", 2 + 3)

# Assigning variables
x = 10
y = 5
z = x * y
print("x * y =", z)

# Writing a simple function
def greet(name):
    return f"Hello, {name}!"

print(greet("Student"))

# Using a for loop
for i in range(5):
    print(f"Number {i}")

# Importing a module and using it
import math
print("Square root of 16 =", math.sqrt(16))
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Basic line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 7]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
